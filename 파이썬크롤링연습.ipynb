```python
import urllib.request as req
from bs4 import BeautifulSoup
import os
import openpyxl
import datetime
from openpyxl.drawing.image import Image
import re

# 이미지 저장할 폴더 생성
if not os.path.exists("./멜론이미지"):
    os.mkdir("./멜론이미지")

header = req.Request("https://www.melon.com/chart/index.htm", headers={"User-Agent":"Mozilla/5.0"})
code = req.urlopen(header)
soup = BeautifulSoup(code, "html.parser")
title = soup.select("div.ellipsis.rank01 > span > a")
name = soup.select("div.ellipsis.rank02 > span")
album = soup.select("div.ellipsis.rank03 > a")
img = soup.select("a.image_typeAll > img")

# 엑셀 파일 생성
if not os.path.exists("./멜론_크롤링.xlsx"):
    openpyxl.Workbook().save("./멜론_크롤링.xlsx")

book = openpyxl.load_workbook("./멜론_크롤링.xlsx")
# 쓸데 없는 시트는 삭제하기
if "Sheet" in book.sheetnames:
    book.remove(book["Sheet"])
sheet = book.create_sheet()
now = datetime.datetime.now()
sheet.title = "{}년 {}월 {}일 {}시 {}분 {}초".format(now.year, now.month, now.day, now.hour, now.minute, now.second)
row_num = 1
# 열 너비 조절
sheet.column_dimensions["A"].width = 15
sheet.column_dimensions["B"].width = 50
sheet.column_dimensions["C"].width = 29
sheet.column_dimensions["D"].width = 47
for i in range(len(title)):
    img_dir = "./멜론이미지"
    img_file_name = img_dir + "/" + re.sub("[\\\/:*?\"<>\|]", " ", title[i].string) + ".png"
    req.urlretrieve(img[i].attrs["src"], img_file_name)
    # 엑셀에 크롤링 결과 출력
    img_for_excel = Image(img_file_name)
    sheet.add_image(img_for_excel, "A{}".format(row_num))
    sheet.cell(row=row_num, column=2).value = title[i].string
    sheet.cell(row=row_num, column=3).value = name[i].text
    sheet.cell(row=row_num, column=4).value = album[i].string
    sheet.row_dimensions[row_num].height = 90 # 행높이 조절.
    book.save("./멜론_크롤링.xlsx")
    print("{}위. {} - {}".format(row_num, title[i].string, name[i].text))
    row_num+=1
```

```python
# 신문 기사 아카이브 크롤링 실습 // selenium 쓰지 않고 BeautifulSoup로 로그인하고 크롤링 하기

# 동아일보아카이브 실습 ID: talingpython, PW: 탈잉파이썬2

# 개발자 도구 >> Network >> Preserve log >> All누르고 로그인
# 파일의 Request Method: get(or post 등)으로 써있음
# get 방식: 서버 요청 시, 요청 메시지를 함께 실어 보낸다 -> 간편하나 보안에 취약함
# post 방식: 요청 메시지를 일종의 택배 상자에 실어 보낸다 -> 보안에 강함

import requests
from bs4 import BeautifulSoup

sess = requests.session() # 세션 만들기 (서버와 내 컴퓨터 사이의 연결고리를 만들기 위해서)
data={"idsave_value":"",
"errorChk":"", 
"gourl": "https%3A%2F%2Fwww.donga.com%2Farchive%2Fnewslibrary%2Fview%3Fymd%3D19600428%26mode%3D19600428%2F0000638580%2F1",
"bid": "talingpython",
"bpw": "xkfdldvkdlTjs2"}
h = {"Referer": "https://secure.donga.com/membership/login.php?gourl=https%3A%2F%2Fwww.donga.com%2Farchive%2Fnewslibrary%2Fview%3Fymd%3D19600428%26mode%3D19600428%2F0000638580%2F1"}
sess.post("https://secure.donga.com/membership/trans_exe.php", headers=h, data=data) # post 요청, payload의 Form data내용 가져오기
# data 가져올 때, Headers에 Request Headers의 Referer도 가져오기. 모르겠다면 걍 Request Headers 다 가져오기

# req.urlopen("https://www.donga.com/archive/newslibrary/view?ymd=19600428&mode=19600428/0000638580/1") # 연결고리가 없어지므로 X

code = sess.get("https://www.donga.com/archive/newslibrary/view?ymd=19600428&mode=19600428/0000638580/1") # req.urlopen이랑 똑같음, 연결고리만 살림
soup = BeautifulSoup(code.text,"html.parser") # sess.get 일때만 code.text 사용해야 함.
title = soup.select("div.list_zone a")
for i in title:
    print("제목: {}".format(i.string))
    print()
    
    # html에서 원하는 url 주소를 얻을 수 없는 경우 Network를 통해서 가져올 수 있다
    # 만약 각각의 url 주소의 규칙이 있다면 for문을 이용하여 간단히 해결 가능
    # 그렇지 않으면 열심히 찾기 위해 노력해야 한다( ㅋㅋㅋ )
    content_num = i.attrs["onclick"].replace("javascript:getNewsArticle('19600428/", "").replace("/1'); return false;", "")
    content_url = "https://www.donga.com/archive/newslibrary/view?idx=19600428%2F{}%2F1".format(content_num)
    code = sess.get(content_url)
    soup = BeautifulSoup(code.text, "html.parser")
    content = soup.select_one("div.article_txt")
    print(content.text.strip())
    print()
```
